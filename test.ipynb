# Яблочная парша #

```python
def apple_parsha(temp, hum):
    percent_growth = []
    index = 0
    for i in range(len(temp)):
        if 70 < hum[i] < 90 and 0 < temp[i] <= 19: # значение может стать больше 100, необходимо исправить
            index = index + temp[i] * 100 / 19 * 24 / 500
            if index >= 100:
                index = 100
            percent_growth.append(round(index, 2))
        else: 
            if 70 < hum[i] < 90 and 19 < temp[i] < 30:
                index = index + (-temp[i] * 3000 / 330 + 3000 / 11) * 24/500
                if index >= 100:
                    index = 100
                percent_growth.append(round(index,2))
            else:
                if hum[i] <= 70 or -3 < temp[i] <= 0 or 30 <= temp[i] < 32: 
                    #index = index1
                    percent_growth.append(round(index, 2))
                else:
                    if hum[i] >= 90 or temp[i] <= -3 or  temp[i] >= 32: 
                        index = 0
                        percent_growth.append(index)
    return percent_growth
```

# Огненная вспышка #

```python
def growth_rate_fire(temp, hum):
    bacterial_count = []
    index = 100
    for i in range(len(temp)):
        if hum[i] >= 90:#проникновение бактерий Erwinia amylovora на рыльце цветка 
            index = 0
            bacterial_count.append(round(index, 2))        
        else: 
            if temp[i] <= 3.9: 
                index = 0
                bacterial_count.append(round(index, 2))
            else: 
                if 3.9 <= temp[i] < 17.22:
                    index = index + (temp[i]*17.6 + 24.5) / (0.91432 * 85.78656)
                    bacterial_count.append(round(index,2))
                else:
                    if  17.22 <= temp[i] < 25.56: 
                        index = index + (temp[i]*176.74 - 2723.46) / (0.91432 * 85.78656)
                        bacterial_count.append(round(index, 2))
                    else:
                        if 25.56 <= temp[i] < 38.71: 
                            index = index + (temp[i]*(-128.6) + 5081) / (0.91432 * 85.78656)
                            bacterial_count.append(round(index, 2))
                        else:
                            if 38.71 <= temp[i] <= 40.13: 
                                index = index + 100 / (0.91432 * 85.78656)
                                bacterial_count.append(round(index, 2))
                            else:
                                if temp[i] > 40.13: 
                                    index = 0
                                    bacterial_count.append(index)
    return bacterial_count
```

# Яблонная плодожорка #

```python
def laying_rate_moth(temp):
    rate_moth = []
    index = 0
    for i in range(len(temp)):
        if temp[i] > 10 and temp[i] <= 17:
            index = temp[i] * 100/7 - 1000/7
            rate_moth.append(round(index,2))
        else: 
            if temp[i] > 17 and temp[i] < 30:
                index = -temp[i] * 100/15 + 640/3
                rate_moth.append(round(index,2))
            else: 
                if temp[i] < 17 or temp[i] > 30:
                    index = 0
                    rate_moth.append(round(index,2))
    return rate_moth
```

```python
df['laying_rate_moth'] = laying_rate_moth(df.t)
```
